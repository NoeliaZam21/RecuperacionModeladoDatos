Es decir, no puede existir en una columna datos con mayúsculas y minúscula de manera simultánea (mezclado).

## **Conexión a la base de datos**


Conexión a la base de datos Northwind_large

```python
path = "sqlite:///DB_transaccional.db"
motorDB = sqlalchemy.create_engine(path)
conectarDB = motorDB.connect()
```

Función del proceso de extracción

```python
def extraer_datos_sql(conectarDB,query):
    result = conectarDB.execute(query)
    df = pd.DataFrame(result.fetchall())
    df.columns = result.keys()
    return df
```

Extracción de datos para la dimension tipo

```python
query = '''SELECT Proposed_Construction_Type,
                Existing_Construction_Type,
                Existing_Construction_Type_Description,
                Proposed_Construction_Type_Description
            FROM Tipo;'''
df_dimTipo = extraer_datos_sql(conectarDB,query)
df_dimTipo.head()
```

Extracción de datos para la dimension Registro

```python
query = '''SELECT Zipcode,
            Location,
            id_Permit,
            id_Tipo
        FROM Registro;'''
df_dimregitro = extraer_datos_sql(conectarDB,query)
df_dimregitro.head()
```

Extracción de datos para la dimension Permisos

```python
query = '''SELECT Permit_Number,
            Permit_Type,
            Permit_Type_Definition,
            Permit_Creation_Date,
            id_street
        FROM Permisos;'''
df_dimpermiso = extraer_datos_sql(conectarDB,query)
df_dimpermiso.head()
```

Extracción de datos para la dimension calle

```python
query = '''SELECT Lot,
            Street_Number,
            Street_Name,
            First_Construction_Document_Date,
            Description
        FROM Calle;'''
df_dimcalle = extraer_datos_sql(conectarDB,query)
df_dimcalle.head()
```

Extracción de datos para la dimension obra

```python
query = '''SELECT Block,
              Neighborhoods_Analysis_Boundaries,
              Filed,
              Supervisor_District,
              Id_record,
              id_Estado
          FROM Obra;'''
df_dimobra = extraer_datos_sql(conectarDB,query)
df_dimobra.head()
```

Extracción de datos para la dimension estado

```python
query = '''SELECT Street_Suffix,
            Current_Status,
            Current_Status_Date,
            Issued_Date,
            Completed_Date
        FROM Estado;'''
df_dimestado= extraer_datos_sql(conectarDB,query)
df_dimestado.head()
```

Extracción de datos para la dimension diseño

```python
query = '''SELECT Unit,
            Number_of_Existing_Stories,
            Fire_Only_Permit,
            id_costo,
            id_Obra
        FROM Disenio;'''
df_dimdisenio= extraer_datos_sql(conectarDB,query)
df_dimdisenio.head()
```
